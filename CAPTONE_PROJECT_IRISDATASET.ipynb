```python
print(f"Min of sepal_length : {np.min(iris_df.sepal_length)}")
print(f"Max of sepal_length : {np.max(iris_df.sepal_length)}")
print(f"Average of sepal_length : {np.average(iris_df.sepal_length)}")
print(f"Median of sepal_length : {np.median(iris_df.sepal_length)}")
print(f"Std of sepal_length : {np.std(iris_df.sepal_length)}")


```

#### Species separately

```python
setosa_df=iris_df[iris_df.species=='setosa']
versicolor_df=iris_df[iris_df.species=='versicolor']
virginica_df=iris_df[iris_df.species=='virginica']
```

```python
print(f"Min of sepal_length in setosa: {np.min(setosa_df.sepal_length)}")
print(f"Max of sepal_length in setosa: {np.max(setosa_df.sepal_length)}")
print(f"Average of sepal_length in setosa : {np.average(setosa_df.sepal_length)}")
print(f"Median of sepal_length  in setosa: {np.median(setosa_df.sepal_length)}")
print(f"Std of sepal_length in setosa : {np.std(setosa_df.sepal_length)}")
```

#### Special_value

```python
def caculator(df):
    special_value=((df.sepal_length *df.sepal_width)-(df.petal_length **df.petal_width)) / (df.sepal_length + df.sepal_width + df.petal_length + df.petal_width)
    df["special_value"]=special_value
    return df
```

```python
new_df=caculator(iris_df)
new_df
```

# Visualization data with Matplotlib and Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df
```

```python
penguins_df=penguins_df.dropna(axis=0)
penguins_df
```

### Matplotlib

##### PEAGUN FEATURE

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,15))
ax1.hist(penguins_df.bill_length_mm,label="nm",color="blue")
ax1.set_title('Penguins bill length')
ax1.set_xlabel('bill_length')
ax2.hist(penguins_df.bill_depth_mm,label="nm",color="green")
ax2.set_title('Penguins bill depth')
ax2.set_xlabel('bill_depth')
ax3.hist(penguins_df.flipper_length_mm,label="nm",color="red")
ax3.set_title('Penguins flipper length')
ax3.set_xlabel('flipper_length')
ax4.hist(penguins_df.body_mass_g,label="g",color="wheat")
ax4.set_title('Penguins body mass')
ax4.set_xlabel('body_mass')
fig.legend()
```

##### The relationship in species, island and sex

```python
species=set(penguins_df.species)
species
```

```python
island=set(penguins_df.island)
island
```

```python
sex=set(penguins_df.sex)
sex
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(nrows=3,ncols=1,figsize=(10,15))

#in species
gr_species1=penguins_df[penguins_df.species=='Adelie']
gr_species2=penguins_df[penguins_df.species=='Chinstrap']
gr_species3=penguins_df[penguins_df.species=='Gentoo']
ax1.scatter(gr_species1.bill_length_mm,gr_species1.bill_depth_mm,label='Adelie');
ax1.scatter(gr_species2.bill_length_mm,gr_species2.bill_depth_mm,label='Chinstrap');
ax1.scatter(gr_species3.bill_length_mm,gr_species3.bill_depth_mm,label='Gentoo');
ax1.set_xlabel('bill_length')
ax1.set_ylabel('bill_depth')
ax1.set_title('The relationship of the two features in species')
ax1.legend()

#in island
gr_island1=penguins_df[penguins_df.island=='Biscoe']
gr_island2=penguins_df[penguins_df.island=='Dream']
gr_island3=penguins_df[penguins_df.island=='Torgersen']
ax2.scatter(gr_island1.bill_length_mm,gr_island1.bill_depth_mm,label='Biscoe');
ax2.scatter(gr_island2.bill_length_mm,gr_island2.bill_depth_mm,label='Dream');
ax2.scatter(gr_island3.bill_length_mm,gr_island3.bill_depth_mm,label='Torgersen');
ax2.set_xlabel('bill_length')
ax2.set_ylabel('bill_depth')
ax2.set_title('The relationship of the two features in island')
ax2.legend()

#in sex
gr_sex1=penguins_df[penguins_df.sex=='Male']
gr_sex2=penguins_df[penguins_df.sex=='Female']
ax3.scatter(gr_sex1.bill_length_mm,gr_sex1.bill_depth_mm,label='Male');
ax3.scatter(gr_sex2.bill_length_mm,gr_sex2.bill_depth_mm,label='Female');
ax3.set_xlabel('bill_length')
ax3.set_ylabel('bill_depth')
ax3.set_title('The relationship of the two features in sex')
ax3.legend()
```

##### The average value of each penguins feature in species , island and sex

```python
gr_species=penguins_df.groupby(by='species').agg({'bill_length_mm':np.average,'bill_depth_mm':np.average,'flipper_length_mm':np.average,'body_mass_g':np.average}).reset_index()
gr_species
```

```python
gr_island=penguins_df.groupby(by='island').agg({'bill_length_mm':np.average,'bill_depth_mm':np.average,'flipper_length_mm':np.average,'body_mass_g':np.average}).reset_index()
gr_island
```

```python
gr_sex=penguins_df.groupby(by='sex').agg({'bill_length_mm':np.average,'bill_depth_mm':np.average,'flipper_length_mm':np.average,'body_mass_g':np.average}).reset_index()
gr_sex
```

```python
X_axis = np.arange(3)
width=0.1

plt.bar(X_axis,gr_species.bill_length_mm,width,label='bill_length')
plt.bar(X_axis+width,gr_species.bill_depth_mm,width,label='bill_depth')
plt.bar(X_axis+width*2,gr_species.flipper_length_mm,width,label='flipper_length')
plt.bar(X_axis+width*3,gr_species.body_mass_g,width,label='body_mass')

plt.xlabel('species')
plt.xticks(X_axis+width, gr_species.species)
plt.title('The average value of each penguins feature in species')
plt.legend()
plt.show()
```

```python
X_axis = np.arange(len(gr_island.island))
width=0.1

plt.bar(X_axis,gr_island.bill_length_mm,width,label='bill_length')
plt.bar(X_axis+width,gr_island.bill_depth_mm,width,label='bill_depth')
plt.bar(X_axis+width*2,gr_island.flipper_length_mm,width,label='flipper_length')
plt.bar(X_axis+width*3,gr_island.body_mass_g,width,label='body_mass')

plt.xlabel('island')
plt.xticks(X_axis+width, gr_island.island)
plt.title('The average value of each penguins feature in species')
plt.legend()
plt.show()
```

```python
X_axis = np.arange(len(gr_sex.sex))
width=0.1

plt.bar(X_axis,gr_sex.bill_length_mm,width,label='bill_length')
plt.bar(X_axis+width,gr_sex.bill_depth_mm,width,label='bill_depth')
plt.bar(X_axis+width*2,gr_sex.flipper_length_mm,width,label='flipper_length')
plt.bar(X_axis+width*3,gr_sex.body_mass_g,width,label='body_mass')

plt.xlabel('sex')
plt.xticks(X_axis+width,gr_sex.sex)
plt.title('The average value of each penguins feature in sex')
plt.legend()
plt.show()
```
